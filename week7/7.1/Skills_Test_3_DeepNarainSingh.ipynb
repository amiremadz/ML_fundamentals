6 assumtions of BLUE are 

1. The model is linear with respect to parameters:
$$Y = \hat{\beta_{0}}+\hat{\beta_{1}}x_{1}+\hat{\beta_{2}}x_{2}+ \cdots +\hat{\beta_{n}}x_{n}$$

2. It is assumed that the training set is a representative random sample. i.e residual
are independent and uncorelated to each other.

3. There is no perfect collinearity between the features. It ensures that $(X^{T}X)^{-1}$ exists.

4. Constant conditional mean. $E(\epsilon|X) = C$ This means that you have properly specified the model such that: there are **no omitted variables**, and the functional form you estimated is correct relative to the (actually unknown) population model. This is always the problematic assumption with OLS, since there is no way to ever know if it is actually valid or not. If $C$ is actually constant but non-zero, then it will be captured by the constant term in the regression.

5. Variance of the error term is constant. It is also called as assumtion of homoskedasticity. This conditional on all $x_{i}$: $Var(\epsilon|X)=\sigma^{2}$ 

6. Error or residual follows normal distribution.

###Derive and write the cost function of OLS

Curve to fit the data is
$f(x) = \beta_0 + \beta_1 x$

We want to determine the coefficients $\beta_0$ and $\beta_1$ which best fits the data
so that the sum of the squares of the vertical errors $e_1, e_2, \cdots, e_n$ is minimal

$$e_i = \left |f(x_i) - y_i \right| = \left |\beta_0 + \beta_1 x_i - y_i \right|$$

We have to find the values of the coefficients that minimizes the following sum:



**So Cost Function is** : $$\sum_{i=1}^{n} e_i^2 = \sum_{i=1}^{n} (\beta_0 + \beta_1 x_i - y_i)^2$$

Minimum value occurs at :

$$\frac{\partial}{\partial \beta_0}\left( \sum_{i=1}^{n} e_i^2 \right)= 2 \sum_{i=1}^{n} (\beta_0 + \beta_1 x_i - y_i) = 0$$

$$\frac{\partial}{\partial \beta_1}\left( \sum_{i=1}^{n} e_i^2 \right)= 2 \sum_{i=1}^{n} (\beta_0 + \beta_1 x_i - y_i) x_i = 0$$

Rearranging the terms produces two equations in the two unknown coefficients:

$$\beta_0 \sum 1 + \beta_1 \sum x_i = \sum y_i$$

$$\beta_0 \sum x_i + \beta_1 \sum x_i^2 = \sum y_i x_i$$

Here all summations are from $i=1$ to $i=n$. In matrix form:

$$\left[ \begin{matrix} \sum 1 & \sum x_i \\ \sum x_i & \sum x_i^2 \end{matrix} \right] \left[ \begin{matrix} \beta_0 \\ \beta_1 \end{matrix} \right] = \left[ \begin{matrix} 1 & 1 & \cdots & 1 \\ x_1 & x_2 & \cdots & x_n \end{matrix} \right] \left[ \begin{matrix} y_1 \\ \vdots \\ y_n \end{matrix} \right]$$

Or

$${\bf A}^T {\bf A \beta} = {\bf A}^T {\bf y}$$

where

${\bf A} = \left[ \begin{matrix} 1 & x_1 \\ 1 & x_2 \\ \vdots & \vdots \\ 1 & x_n \end{matrix} \right]$, ${\bf y} = \left[ \begin{matrix} y_1 \\ y_2 \\ \vdots \\ y_n \end{matrix} \right]$, and ${\bf \beta} = \left[ \begin{matrix} \beta_0 \\ \beta_1 \end{matrix} \right]$


The normal equation has a unique solution given by:

$${\bf \beta} = ({\bf A}^T {\bf A})^{-1} {\bf A}^T {\bf y}$$

This minimizes the total sum of squares of the errors given by:

$$\sum_{i=1}^{n} e_i^2 = \sum_{i=1}^{n} (\beta_0 + \beta_1 x_i - y_i)^2 = ({\bf A \beta} - {\bf y})^T ({\bf A \beta} - {\bf y})$$

$${\bf \beta} = ({\bf A}^T {\bf A})^{-1} {\bf A}^T {\bf y}$$

###Describe Lp Norms in the Context of Regularization

Lp Norms can be given by the below equation

$$ \|x\|_p = \left(|x_1|^p+|x_2|^p+\dotsb+|x_n|^p\right)^{\frac{1}{p}} $$

We use Regularisation in order to reduce the overfitting of the model.
In regularisation we shrink the size of the coefficient $\beta$ 
to make the regression model more robust.
Smaller coefficient produce more robust regression as they have less bias and variance.

One way of doing this is to constrain the p-norm of the β coefficients to a maximum possible size.
It is a penalizing technique to improve the performance of  regression model.

So we have different regularisation like

$L^{1}$ Regularisation(LASSO) : We add $L^{1}$ p norm to least square equation and check for the values of $\beta$ that satisfy it. We minimize the below cost function.

$$F(\textbf{$\beta$}) = \sum\limits_{i=1}^n(y_i-\beta_0-\beta_iX)+\lambda \sum\limits_{i=1}^n|\beta_i|$$ 

$L^{2}$ Regularisation (RIDGE Regression)

$$F(\textbf{$\beta$}) = \sum\limits_{i=1}^n(y_i-\beta_0-\beta_iX)+\lambda \sum\limits_{i=1}^n\beta_i^{2}$$ 

$L^{3/2}$ Regularisation (Elastic Net Regression): We add both $L^{1}$ p norm and $L^{2}$ p norm 
and minimize the below cost function

$$F(\textbf{$\beta$}) = \sum\limits_{i=1}^n(\beta_iX-y_i-b_0)+\lambda_2 \sum\limits_{i=1}^n\beta_i^{2}+\lambda_1 \sum\limits_{i=1}^n|\beta_i|$$

### Write the cost function of SVMs

The total **margin** between the support vectors is given by M:

$$M = \frac{2}{\|\beta\|}$$

We haveto maximize the margin.

we are going to minimize $\|\beta\|_2$ subject to our principle equation, $\sum_{i=0}^{n}y_i(\beta_0 + \beta^{T}x_{i}) \geq 1$ .

Cost Function is below(Dual Langrangian):

$$\Lambda(x,y,\alpha) = \sum_{i=0}^{n}\alpha_{i}-\frac{1}{2}\sum_{i=0}^{n}\sum_{j=0}^{n}\alpha_{i}\alpha_{j}y_{i}y_{j}x_{i}^{T}x_{j}$$

subject to some $\alpha_{i} \geq 0$ and $\sum_{i=0}^{n}\alpha_{i}y_{i} = 0$

###Write a description of the DBSCAN clustering algorithm. Describe use cases, weaknesses and strengths.

DBSCAN is data clustering algorithm based on density based approach.
Density-based spatial clustering of applications with noise .
Density-based approaches like DBSCAN model clusters as high-density clumps of points. 

Algorithm is below:
    
    
a)Pick an arbitrary point from the dataset. This is marked as a visited point.
b)Count all points within ϵ of the original point. This is the neighborhood.
c)If there are more than minPoints points within a distance of ϵ from that point
(including the original point itself), label them all as members of a new cluster.
d)For all points in the neighborhood of every point in the cluster, check to see if they also have minPoints within an ϵ neighborhood. Those that are, are also cluster members.
e)If a point has been visited by the algorithm and doesn't belong to a cluster yet, it is marked as noise.

Strength:
    
a)It is fast
b)DBSCAN is resistant to noise and can handle clusters of various shapes and sizes
c)Finds non linearly separable clusters
d)Robust to outliers
e)O(NlogN) is the time complexity of the algorithm.
f)We dont need to specify the no of clusters as in K-means.
    
Weakness:
    
a)DBSCAN does not work to well when we are dealing 
    with clusters of varying densities or with high dimensional data.
    
b)It relies heavily on distance measure (Eucilidean distance). In higher dimension hard
to find good value of epsilon.

c)Curse of Dimensionality

d)Border points that are reachable from more than one cluster can be part of either cluster
depending on the order data is processed.

###Describe the standard metrics of description in clustering: Dispersion, Gap and Silhouette.

Dispersion:
---------------
A  metric to tell how good clustering alorgithm is performing is the within-cluster dispersion.

$$D_{r} = \sum_{i, i' \in C_{r}} d_{ii'}$$

$$W_{k} = \sum_{r = 1}^{k} \frac{1}{2n_{r}}D_{r}$$

As $k$ increases $W_{k}$ monotonically decreases.  So the optimal $k$ is when each data point is its own cluster center which is not quite useful so we determine the smallest number of clusters $k$ with the lowest error.

Graphical method to determine is the elbow plot.
Is is used to determine when we are getting diminishing returns for every increase in $k$.  This technique looks for a kink in the error plot.

Gap Statistic:
-----------------------------
    
The Gap Statistic is another technique to estimate the optimal number of clusters in a data set.  It  uses the data points, and also compares them to a synthetic "null" distribution or random noise.  The Gap statistic can be thought of an automated way to determine the "kink" or "elbow" in the plot of the dispersion.

Also, one of the most apparent weakness of the Silhouette (and most other metrics) is when the true data has no discernible clusters or only one true cluster. The Gap statistic solves some of the issues with the Silhouette by introducing the concept of a reference null dataset to compare performance against.  

    

Silhouette : 
----------------------
It measures the quality of clustering.
    It determines how well each object lies within the cluster.
    It improves  the cluster dispersion by also accounting for how similar each point is to every 
    other point in its cluster instead of just taking the centroid.
    A high silhouette width indicates a good clustering.
    
    Silhouette metric ranges 
    from $-1 \leq s(i) \geq 1$
    
    In comparison to the elbow method, the silhouette does not monotonically decrease as $k$  increases.  Because of this we can simply vary $k$ and look for when the silhouette is at a maximum.

```python
## Adapted from Scikit learn implementation
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

from sklearn.utils import shuffle
from sklearn.utils import check_random_state
from sklearn.cluster import MiniBatchKMeans
from sklearn.cluster import KMeans

random_state = np.random.RandomState(0)

# Number of run (with randomly generated dataset) for each strategy so as
# to be able to compute an estimate of the standard deviation
n_runs = 5

# k-means models can do several random inits so as to be able to trade
# CPU time for convergence robustness
n_init_range = np.array([1, 5, 10, 15, 20])

# Datasets generation parameters
n_samples_per_center = 100
grid_size = 3
scale = 0.1
n_clusters = grid_size ** 2


def make_data(random_state, n_samples_per_center, grid_size, scale):
    random_state = check_random_state(random_state)
    centers = np.array([[i, j]
                        for i in range(grid_size)
                        for j in range(grid_size)])
    n_clusters_true, n_features = centers.shape

    noise = random_state.normal(
        scale=scale, size=(n_samples_per_center, centers.shape[1]))

    X = np.concatenate([c + noise for c in centers])
    y = np.concatenate([[i] * n_samples_per_center
                        for i in range(n_clusters_true)])
    return shuffle(X, y, random_state=random_state)

# Part 1: Quantitative evaluation of various init methods

fig = plt.figure()
plots = []
legends = []

cases = [
    (KMeans, 'k-means++', {}),
    (KMeans, 'random', {}),
    (MiniBatchKMeans, 'k-means++', {'max_no_improvement': 3}),
    (MiniBatchKMeans, 'random', {'max_no_improvement': 3, 'init_size': 500}),
]

for factory, init, params in cases:
    print("Evaluation of %s with %s init" % (factory.__name__, init))
    inertia = np.empty((len(n_init_range), n_runs))

    for run_id in range(n_runs):
        X, y = make_data(run_id, n_samples_per_center, grid_size, scale)
        for i, n_init in enumerate(n_init_range):
            km = factory(n_clusters=n_clusters, init=init, random_state=run_id,
                         n_init=n_init, **params).fit(X)
            inertia[i, run_id] = km.inertia_
    p = plt.errorbar(n_init_range, inertia.mean(axis=1), inertia.std(axis=1))
    plots.append(p[0])
    legends.append("%s with %s init" % (factory.__name__, init))

plt.xlabel('n_init')
plt.ylabel('inertia')
plt.legend(plots, legends)
plt.title("Mean inertia for various k-means init across %d runs" % n_runs)

# Part 2: Qualitative visual inspection of the convergence

X, y = make_data(random_state, n_samples_per_center, grid_size, scale)
km = MiniBatchKMeans(n_clusters=n_clusters, init='random', n_init=1,
                     random_state=random_state).fit(X)

fig = plt.figure()
for k in range(n_clusters):
    my_members = km.labels_ == k
    color = cm.spectral(float(k) / n_clusters, 1)
    plt.plot(X[my_members, 0], X[my_members, 1], 'o', marker='.', c=color)
    cluster_center = km.cluster_centers_[k]
    plt.plot(cluster_center[0], cluster_center[1], 'o',
             markerfacecolor=color, markeredgecolor='k', markersize=6)
    plt.title("Example cluster allocation with a single random init\n"
              "with MiniBatchKMeans")

plt.show()
```

```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
 
class DBSCANClass:
    
            
        
    #    Points P within eps neighborhood 
    
    def regionQuery(self,P,eps,neighborhoods):
        return neighborhoods[P]
        
     
    #Computing eucilidean distnace
    def Distance(self,X):
        
        return euclidean_distances(X)
    
    


    def DBSCAN(self, D , eps, MinPts):
        
        #Raise error if eps <0
        
        if not eps > 0.0:
            raise ValueError("eps must be positive.")
            
        #Number of points
        
        numofPoints = D.shape[0]
        
        #Retrieving Eucilidean distance matrix
        
        dist = self.Distance(D)
        
        #find all neighbourhoods within eps distance
        neighborhoods = [np.where(x<=eps)[0] for x in dist]

        C = 0
        
        visited = [False]* numofPoints
        
        # initialise labels to all -1
        labels = -np.ones(numofPoints)
        
        # for each point P in dataset D
        for p in range(0,numofPoints):
            # if P is visited
            if visited[p] == True:
                
                # move to next point
                continue
                
            # mark P as visited
            visited[p] = True
            
            #Neighbourhodd points
            
            NeighbourPts = self.regionQuery(p,eps,neighborhoods)

            # if sizeof(NeighborPts) < MinPts
            
            if len(NeighbourPts) < MinPts:
                # mark P as NOISE
                labels[p] = -1
            else:
                # Increment the cluster value
                C += 1
                
                self.expandCluster(p,NeighbourPts,C,eps,MinPts,visited,labels,neighborhoods)
        return labels
    
    def fit(self,D , eps, MinPts):
        return self.DBSCAN(D , eps, MinPts)
    
    
    def expandCluster(self, p, NeighborPts, C, eps, MinPts, visited, labels, neighborhoods):
        
        #Adding point to cluseter
        
        labels[p] = C
        i = 0
        while i < len(NeighborPts):
            neighbourp = NeighborPts[i]
            if visited[neighbourp] == False:
                visited[neighbourp] = True
                NeighborPts2 = self.regionQuery(neighbourp,eps,neighborhoods)
                if len(NeighborPts2)>=MinPts:
                    NeighborPts = np.append(NeighborPts,NeighborPts2)
            if labels[neighbourp] == None or labels[neighbourp]==-1:
                labels[neighbourp] = C
            i +=1

    def predict(self, D , eps, MinPts):
        self.fit(D , eps, MinPts)
        return self.labels
        
```
